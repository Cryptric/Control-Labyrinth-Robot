```python
import sys
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
file = "../RecordTest/img-20.jpg"
img = cv.imread(file, cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
blur_img = cv.medianBlur(img, 5)
plt.imshow(blur_img, cmap='gray')
```

```python
rows = blur_img.shape[0]
circles = cv.HoughCircles(blur_img, cv.HOUGH_GRADIENT, 1, 16, param1=1, param2=20, minRadius=3, maxRadius=30)
plt.imshow(img, cmap='gray')
plt.scatter(circles[0, :, 0], circles[0, :, 1])
```

```python
pos = np.unravel_index(img.argmax(), img.shape)
plt.imshow(img, cmap='gray')
plt.scatter(pos[1], pos[0])
```

```python
pattern = img[12:20, 182:190]
plt.imshow(pattern, cmap='gray')
```

```python
res = cv.matchTemplate(img, pattern, cv.TM_CCOEFF)
plt.imshow(res)
```
